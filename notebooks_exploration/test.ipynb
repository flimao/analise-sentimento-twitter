```python
import numpy as np
import pandas as pd
from mltoolkit import metrics, plots

from matplotlib import rcParamsDefault, rcParams, pyplot as plt

rcParams.update(rcParamsDefault)
rcParams['figure.dpi'] = 120
rcParams['figure.figsize'] = (10, 8)
```

```python
df = pd.read_csv(r'../../../datasets/college.csv')
```

```python
X = df.drop(columns = ['Name'])
y = df['Grad.Rate']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.3
)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(random_state=42)
reg.fit(X_train, y_train)
```

```python
plots.plot_true_pred(
    modelos = { 'RF': reg },
    X_test = X_test, y_test = y_test
);
```

```python
df = pd.read_csv(r'../../../datasets/winequality-red.csv')
df.head()
```

```python
X = df.drop(columns = ['quality'])
y = pd.cut(df.quality, bins = [0, 7, 10], labels = [0, 1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.3,
    stratify = y
)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
plots.plot_roc_auc_curve(
    models = { 'RF': clf },
    X_test = X_test, y_test = y_test
)
```

```python
plots.plot_precision_recall_curve(
    y_true = y_test,
    y_prob = clf.predict_proba(X_test)[:, 1]
);
```

```python
plots.plot_threshold_curve(
    y_true = y_test,
    y_prob = clf.predict_proba(X_test)[:, 1],
    metrics = {
        'f1_score': 'F1',
        'precision': 'Precision'
    }
);
```
